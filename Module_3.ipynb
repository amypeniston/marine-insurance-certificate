#### Introduction

The most significant change was the replacement of the S and G form in the early 1980s with the new MAR type of contract with attached clauses. These new clauses were, first and foremost, better written. They also improved ship valuation by inserting the disbursements clause, which limited “extraneous” items that might otherwise be included in the value of the ship.

#### Types of Policies

* Voyage policies: coverage for one specific voyage or group of voyages.
* Time policies: typical coverage is for one year, with a “continuation clause” to protect in the instance that the policy expires while the ship is at sea – this extends coverage in return for pro-rata premium until the vessel arrives at the next port (subject to prior UW approval).
* Construction policies aka Builders’ Risks: cover against damage during construction or extensive repairs. May cover the vessel and machinery.
* Port policies: typically cover extended lay-up periods.
* Interest policies: covers financial interest such as disbursements.
* Honour policies: otherwise known as Policy Proof of Interest Policies aka Gambling policies. The insured’s insurable interest is “admitted” even though none may actually exist. If governed by English Law, these policies have no legal standing. Under other legal systems, the burden is on the insured to prove insurable interest.
* Fleet policies: cover more than one vessel operating under the same ownership.

#### Time and Voyage Policies

Until 1982, most hull and cargo insurances were “voyage” policies. However, since then, a practice has developed so that most ships are now insured using a “time” policy. Cargo “voyage” policies have been extended to cover from “warehouse to warehouse”.

The distinction between time and voyage insurances is important because:

* Voyage policies are subject to the warranty of seaworthiness and have an implied condition that the ship must sail within a reasonable time after the policy is written or the underwriter can avoid the contract; the voyage must commence from the named port and proceed in a customary fashion otherwise cover does not attach. If the destination is changed after departure, coverage attaches, but only continues until the “change in voyage” decision was made. However, in practice, there is a hold covered condition whereby a change in voyage is permitted subject to prompt advice to the underwriters and payment of additional premium. If the ship deviates the course, the underwriter remains liable only until the time of deviation, even if the ship regains proper course; however deviations can also be held cover with special wording. Unlike change in voyage and deviation, unreasonable delays are not held covered by standard hull voyage clauses but are instead permitted only in certain circumstances:
  * If authorized by special policy wording
  * Where caused by circumstances beyond the control of the master and his employer
  * Where necessary to comply with a warranty or for the safety of the ship or subject matter insured
  * To save a life or obtain medical assistance
  * When caused by barratrous conduct, if barratry is an insured peril
  * Time policies are subject to less strict seaworthiness warranties

Time and voyage insurances can be included in the same policy, but it is unusual and results in difficult court decisions, as was the case in The AI Jabail IV.

#### Institute Time Clauses (Hull) – 1/11/1995

Clause 1: Navigation
Clause 1.1 aka “Tow and Assist wording” – ensure that vessels are not covered when put to unusual use without prior agreement
Clause 1.2 – shipowners agrees to indemnify pilot/tug against any liability incurred 
Clause 1.3 – allows for transportation of supplies/equipment/personnel (not cargo, unless otherwise approved) to the vessel by helicopter
Clause 1.4 – does not cover transfer of cargo between vessels at sea, unless otherwise agreed
Clause 1.5 – protects underwriters in the event a ship is sold for scrap at a lower value than stated in the policy

Clause 2: Continuation
Allows for the continuation of coverage if the vessel is at sea at the time that the policy expires, subject to prior notice and additional monthly pro-rata premium.

Clause 3: Breach of Warranty
A shipowner is permitted to break certain warranties given he provides immediate notice to the underwriters.

Clause 4: Classification
Clause 4.1 – the vessel must be classed with a Classification Society and maintain said classification throughout the period of insurance.
Clause 4.2 – affirms that Clause 4.1 is a warranty with breach having the same ramifications as any other warranty.
Clause 4.3 – anything potentially causing re-classification must be reported immediately to the Classification Society.
Clause 4.4 – the assured will authorize the underwriter’s request for more information from the Classification Society.

Clause 5: Termination
Clause 5.1 – insurance will be cancelled automatically in the event of any circumstances named in the clause, with pro-rata daily premium returned, i.e. if classification society/class is changed or periodic surveys become overdue.
Clause 5.2 – significant factors that result in termination include change in ownership/management, flag, bareboat charter and requisition for title/use.

Clause 6: Perils
Clause 6.1 – covered perils include “perils of the seas, rivers, lakes or other navigable waters”, where “perils of the sea” are a holdover from the SG form of old. Only fortuitous events are covered, which does not include collision third party liability or the ordinary action of the wind and waves. 6.1.2 – to claim under the peril of fire, the insured need only show loss by fire; the burden is on the insurer to show deliberate act. 6.1.3 – replace “assailing thieves” in SG form with “violent theft”. 6.1.4 – covers loss/damage due to jettison, or the throwing overboard of items to save the ship. 6.1.5 – covers piracy (including on territorial waters, not only the high seas as is international law). 6.1.6 – covers low by contact with docks. 6.1.7 – covers loss due to “acts of God”, i.e. earthquake, volcanic eruption or lightning. 6.1.8 – covers accidents in loading, discharging or shifting of cargo/fuel.
Clause 6.2 – stems from the Inchmaree case in 1887. 6.2.1 – coverage is provided for damage caused by breakage of shafts/boilers, unless due to negligence and subject to the due diligence proviso (i.e. there was no negligence). Cover for the damaged parts can be purchased for additional premium under the Additional Perils Clause. 6.2.2 – covers loss due to negligence of the masters/officers/crew/pilots. 6.2.3 – covers negligence of repairers or charterers. 6.2.4 – covers barratry of the master/officers crew, i.e. willfully committed wrongful acts committed to the prejudice of the owner/charterer. 6.2.5 – covers damage due to aircrafts, helicopters or objects falling therefrom.

Clause 7: Pollution Hazard
Covers damage to the vessel caused by liability, not pollution liability (which is available via P&I clubs).

Clause 8: 3/4ths Collision Liability
Clause 8.1 – Outlines coverage for three fourths of any sum paid by the assured in respect of a collision for which they are liable. Clause 8.1.1 – establishes that covered amounts will be in respect of damage to the other vessel, rather than property or injury. Clause 8.1.2 – covers delay to or loss of use resulting from collision. Clause 8.1.3 – covers General Average or salvage of the other vessel. 
Clause 8.2 – Stipulates that indemnity provided under Clause 8 is additional to other clauses in ITC Hulls. Clause 8.2.1 – indemnity shall be calculated according to principle of cross-liabilities. Clause 8.2.2 – limits liability to 3/4th of the insured value of the insured vessel in any one collision. Clause 8.2.3 – 3/4th of the legal costs incurred will be paid, with prior underwriter consent.  

Clause 9: Sistership
Clause 9 follows from the fact that a person cannot be legally liable to himself.

Clause 10: General Average and Salvage 
While the Marine Insurance Act 1906 treats salvage and general average as separate actions, they can overlap, for example in Rule VI of the York-Antwerp Rules of 1974. The general theme is that the actions were undertaken to preserve property that was imperiled on the common adventure. 
Clause 10.1 – underwriters cover the vessel’s proportion of charges. Clause 10.2 – adjustment of general average will be governed by the law of the place where the adventure ends. Clause 10.4 – avoids indemnifying claims where loss is not incurred to avoid an insured peril. Clause 10.5 – reflects changes in law due to the Salvage Convention of 1989, particularly with respect to preventing environmental damage.
Clause 11: Duty of Assured (Sue and Labor)
An additional sum is recoverable in respect to measure taken to minimize damage in the event of imminent loss. A ransom paid to pirates is not recoverable as general average but can be recoverable as a sue and labour expense (if piracy is a covered peril).

Clause 13: Notice of Claim and Tenders
Clause 13.1 – requires underwriters be notified of loss within 12 months so that they can appoint a surveyor. Clause 13.2 – gives underwriters the right to choose which port and ultimately which repairer shall be used. Clause 13.3 – allows underwriters to accept tenders for the repairs.

Clause 14: New for Old
Underwriters agree not to make deductions when replacing old parts with new parts.

Clause 16: Wages and Maintenance
Restricts indemnity for wages and maintenance.

Clause 18: Unrepaired Damage
Outlines the indemnity paid for claims for unrepaired damages.

Clause 19: Constructive Total Loss
Constructive total losses are total losses in law but not in fact and must be accompanied by a notice of abandonment. The theft of cargo/ship by pirates is unlikely to be considered a total loss of any kind. Clause 19 stipulates that the insured value, rather than market value, is taken as the repaired value for the purposes of determining whether a constructive total loss has occurred. It also outlines how charges are handled in the event of sequential damages.

Clause 20: Freight Waiver
Upon notice of abandonment, the insurer is entitled to property and any freight in course of being earned. 

Clause 21: Assignment
Includes rules for the assignment of the policy to another assured.

Clause 23: Returns for Lay-Up and Cancellation

Clause 23.1.1 – allows for pro-rata monthly returned premium in the even the vessel is no longer subject to navigating and trading perils (i.e. it is laid up). Clause 23.2.3 – no returns of premium are given whilst vessel is used for storage of cargo or lightering purposes.

Clause 24: War Exclusion
The first of four clauses appearing under the wording “the following clauses shall be paramount and shall override anything contained in this insurance inconsistent therewith”. Damage due to “derelict weapons of war” is not recoverable.

Clause 25: Strikes Exclusion
War and strikes perils are covered in the Institute War and Strikes Clauses Hulls 1/11/95. Piracy and violent theft toe the line of being excluded, depending on the circumstances.

Clause 26: Malicious Acts Exclusion
Excludes malicious acts due to the detonation of an explosive or weapon of war or caused by any person acting with a political motive.

Clause 27: Radioactive Contamination Exclusion
A comprehensive clause that excludes all loss, damage or expense directly or indirectly caused by radioactive contamination. 

#### Institute Voyage Clauses (Hulls) – 1/11/1995

“Where the contract is to insure the subject-matter at and from, or from one place to another or others, the policy is called a voyage policy”. The use of “at” denotes coverage whilst the ship is in port waiting to sail, including loading. The use of “from” denotes that coverage is to begin at the point that the ship actually sails. 

Clause 2 “Change of Voyage” – substitutes Clause 2 “Continuation” of the Time Clauses. It allows deviation to be held covered with immediate notice.

#### Proximate Cause

The insurer is liable to indemnify the insured in respect of damages proximately caused by a covered peril. Proximate refers to proximate in efficiency and not necessarily in time; typically only one cause is found to be proximate cause of loss. However, in rare cases where there are two proximate causes and only one is within the terms of the policy, the insurer is obligated to pay unless an exclusion applies to the second proximate cause.

#### Hull Underwriting

Factors contributing to determination of rate:

* Trade
* Flag
* Ownership/Management
* Classification
* Claims experience
* Market competition

In assessing the risk, the underwriter makes a judgement on:

* Level of claims experienced by the underwriter for that type of vessel and its trade
* Valuation requested by the owners – extremely high or extremely low valuations suggests potential moral hazard
* Costs of repair in specific geographic region
* Potential for total loss vs. particular average (accidental partial loss)
* Collision liability (“Running Down Clause”), which is supplementary to the main hull contract
* General Average contributions
* Salvage charges
* Sue and Labour charges, for which cover is supplementary to the main hull contract

A bulk premium (ex. Total loss), which covers partial loss risks plus profit margin, is quoted as a figure in dollars multiplied by dead-weight tonnage, or the maximum weight a ship can carry. This is expressed as a percent of insured value. To this rate, an additional rate is added for the total loss aspect.

While dead-weight tonnage is used for large vessels because it gives a reliable indication of the size and earning potential, for smaller vessels, rates are often based on the power of the engines. For cruise/passenger vessels, the gross register tonnage (GRT, 100 cubic feet = 1 ton) is used, while LNG/LP carriers are rate based on cubic meters.

Partial loss rate varies by trade. For example, ferries are at higher risk for collision due to the frequency of docking and undocking.

Deductibles reduce small, inevitable claims and serve to give the owner financial incentive to ensure the safe outcome of any voyage. This is referred to as “risk improvement”.
Geographic trade regions impact pricing. The main global trading areas are included in the Institute Warranties; the underwriter may make allowances for limited trading outside these regions, subject to additional premiums.

Some underwriters specialize in total loss only conditions (LTO), where rates are determined by vessel characteristics and loss history. A proper valuation is critical, as too low a value may invite moral hazard. Restricted conditions may be added to hull coverage that result in relative benefits in rate. Likewise, ship values may change (usually in the form of depreciation), which will require reconsideration by the underwriter. 

Long-term policies are viewed favorably by underwriters, as they represent a constant income stream and typically mean a well-taken-care-of ship. Likewise, mortgagees often require a “letter of undertaking” from the brokers involved to guarantee that the terms, conditions and, importantly, valuation, remain unchanged.

#### Ship Valuation

In addition to the intrinsic value of the ship, the Disbursements Warranty in the Hull Clauses (1995) sets out items that can contribute to the valuation:
* Disbursement, manager’s commission, increased value, profit: not to exceed 25% of vessel value
* Freight: not to exceed 25% of vessel value
* Time charter: not to exceed 50% of the gross hire earned under charter less than 18 months
* Insurance premiums


#### Claims
The Institute Hull Clauses follow a “named perils” approach. If a claim is made and proof provided, the burden shifts to the insurer to show that the claim is not covered due to an exclusion.
Excluded:
* Any loss not proximately caused by a peril insured against
* Any loss attributable to willful misconduct of the assured – for example, scuttling of the vessel
* Any loss proximately caused by delay
* Ordinary wear and tear, inherent vice etc. – unless otherwise stated in the policy
